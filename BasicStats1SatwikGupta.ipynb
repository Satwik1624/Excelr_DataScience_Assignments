```python
import pandas as pd
```

```python
df = pd.read_csv("sales_data_with_discounts.csv")
```

# Analysis of Histogram

1. **Volume**

  * Outliers: The few instances of higher volumes (above 20) can be considered outliers since the majority are clustered around lower values.
2.  **Avg Price**
  * Distribution: The majority of the average prices are concentrated at the lower end (close to 0), with a small number of transactions having significantly higher prices.
  * Skewness: This distribution is also right-skewed, with a heavy concentration at lower values and a long tail stretching toward higher prices.
  * Outliers: There are notable spikes around 50,000-60,000, indicating some high-priced transactions that act as outliers.
3.  **Total Sales Value**
  * Distribution: The total sales value shows a similar pattern as Avg Price, where most transactions have smaller total sales values, and a few have extremely high values.
  * Skewness: Highly right-skewed with a long tail extending toward higher total sales values.
  * Outliers: A few transactions have sales values reaching up to 200,000, far beyond the majority, indicating strong outliers.
4.  **Discount Rate(%)**
  * Distribution: The discount rate is distributed more evenly across a range, but most transactions have a discount rate between 15-20%
  * Skewness: This appears more left-skewed, with most of the discounts falling in the higher range and fewer transactions having smaller discount rates.
  * Outliers: There don't appear to be significant outliers in this case; most discounts are clustered between 15-20%, which seems reasonable for typical discounting.
5. **Discount Amount
  * Distribution: Most of the discount amounts are clustered near 0-2,500, with a few very high values.
  * Skewness: Right-skewed with a long tail extending toward larger discount amounts.
  * Outliers: The few very large discount amounts (over 20,000) act as clear outliers.
6.  **Net Sales Value**
  * Distribution: Similar to Total Sales Value, most transactions have smaller net sales values, and there are some very high net sales values in the dataset.
  * Skewness: Right-skewed with a long tail toward higher values.
  * Outliers: The high net sales values reaching up to 175,000 are outliers, as most of the data is concentrated in the lower range.

# Analysis of Boxplots
Boxplots provide a visual summary of the distribution, central tendency, and variability of data, while also highlighting outliers (points that lie outside the whiskers of the boxplot)


1.   **Volume**

  *   Outliers: Several outliers are present, primarily above a volume of 10, with some extreme outliers exceeding 30.
  *   Interquartile Range (IQR): The IQR (middle 50% of the data) ranges between 2 and 6. The median is around 4.
  *   Distribution: The bulk of the data lies between 2 and 8, but the outliers suggest there are instances where the volume is significantly higher.
2.   **Avg Price**

  *   Outliers: There are numerous outliers above 30,000, with extreme cases up to 60,000.
  *   IQR: Most of the data lies between 0 and around 10,000. The median is fairly low, indicating that most of the transactions are small, while a few larger transactions inflate the average.
  *   Distribution: The presence of many outliers suggests some high-price sales that differ greatly from the rest of the dataset.
3.   **Total Sales Value**

  *   Outliers: Numerous extreme outliers are present, with total sales values exceeding 125,000 and some going up to 200,000.
  *   IQR: The middle 50% of the data ranges from around 0 to 25,000. The median is close to the lower end, which, combined with the outliers, suggests that most transactions are of smaller sales values with a few high-value transactions.
  *   Distribution: The high number of outliers indicates a right-skewed distribution where a few high sales values are affecting the overall spread.
4. **Discount Rate(%)**
  * Outliers: There are several outliers below 10%. The majority of discounts seem to be above 10%, and the outliers show lower discount rates.
  * IQR: The middle 50% of the discount rates lie between 14% and 19%. The median is around 16%.
  * Distribution: Most transactions have relatively high discount rates, but a few have notably low discounts, which show up as outliers.
5.  **Discount Amount**
   * Outliers: There are a few outliers, with the discount amount exceeding 20,000, which are unusual compared to the bulk of the data.
   * IQR: The bulk of the discounts range between 0 and 2,500, with a median close to 1,000.
   * Distribution: There is a clear distinction between the majority of discount amounts, which are relatively low, and the outliers, which are much larger discounts.

6.  **Net Sales Value**
  * Outliers: Like Total Sales Value, there are many extreme outliers, with values exceeding 100,000 and some going up to 175,000.
  * IQR: Most of the data lies between 0 and 25,000, with a low median. Similar to Total Sales Value, this suggests that a few very large transactions have driven up the overall distribution.
  * Distribution: A right-skewed distribution with significant outliers at the high end of the net sales values.


```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

for col in categorical_cols:
  plt.figure(figsize=(10, 5))
  df[col].value_counts().plot(kind='bar')
  plt.title(f'Bar Chart for {col}')
  plt.show()
```

Standardization (z-score normalization) transforms data so that it has a mean of 0 and a standard deviation of 1. The formula is: z = x-/

Where:

*   x is the value of a data point,
*    is the mean, and

*    is the standard deviation

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standardized = pd.DataFrame(scaler.fit_transform(df[numerical_cols]), columns=numerical_cols)

print("Before Standardization:\n", df[numerical_cols].head())
print("After Standardization:\n", df_standardized.head())
```

One-hot encoding converts categorical columns into binary columns (0 or 1), which are required for most machine learning models.

```python
df.encoded = pd.get_dummies(df, columns=categorical_cols)
print(df.encoded.head())
```
