#Interview Questions:
1. What are some common hyperparameters of decision tree models, and how do they affect the model's performance?

  Ans. Common hyperparameters in decision tree models include max depth, min samples split, min samples leaf, max features, criterion, max leaf nodes, and splitter. These parameters control the tree's complexity, affecting performance by balancing overfitting and underfitting. Proper tuning is essential for enhancing model accuracy and generalization to unseen data.

2. What is the difference between the Label encoding and One-hot encoding?

  Ans. Label encoding and one-hot encoding are methods for converting categorical variables into numerical format. Label encoding assigns a unique integer to each category (e.g., "Red" as 0, "Green" as 1), which can create misleading ordinal relationships. In contrast, one-hot encoding creates separate binary columns for each category (e.g., [1, 0, 0] for "Red"), avoiding false relationships but potentially increasing dimensionality. Label encoding is suitable for ordinal data, while one-hot encoding is better for nominal data.
